## 导入工具包

```python
from mmocr.apis import MMOCRInferencer

import cv2

import matplotlib.pyplot as plt
%matplotlib inline
```

## 载入模型，实例化MMOCRInferencer

SVTR模型库：https://github.com/open-mmlab/mmocr/tree/main/configs/textrecog/svtr

```python
infer = MMOCRInferencer(rec='svtr-small')
```

## 载入预测图像

```python
img_path = 'mmocr/demo/demo_text_recog.jpg'
```

```python
img_bgr = cv2.imread(img_path)
```

```python
plt.imshow(img_bgr[:,:,::-1])
plt.show()
```

## 执行预测

```python
result = infer(img_path, save_vis=True, return_vis=True)
```

## 解析预测结果-文本内容及置信度
