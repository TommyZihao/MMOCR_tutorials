## 导入工具包

```python
from mmocr.apis import MMOCRInferencer

import cv2
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

## 载入模型，实例化MMOCRInferencer

```python
# infer = MMOCRInferencer(det='dbnetpp', rec='svtr-small')

infer = MMOCRInferencer(det='textsnake', rec='svtr-small')
```

## 载入预测图像

```python
img_path = 'mmocr/demo/demo_text_ocr.jpg'

# img_path = 'mmocr/demo/demo_text_det.jpg'
```

```python
img_bgr = cv2.imread(img_path)
```

```python
plt.imshow(img_bgr[:,:,::-1])
plt.show()
```

## 执行预测

```python
result = infer(img_path, save_vis=True, return_vis=True)
```
