## 进入MMOCR主目录

```python
import os
os.chdir('mmocr')
```

## 导入工具包

```python
from mmocr.apis import TextRecInferencer

import matplotlib.pyplot as plt
%matplotlib inline

import cv2
```

## 载入模型

```python
ckpt_path = "work_dirs/sar_resnet31_parallel-decoder_5e_toy/epoch_100.pth"
config_path = "configs/textrecog/sar/sar_resnet31_parallel-decoder_5e_toy.py"
```

```python
infer = TextRecInferencer(config_path, ckpt_path)
```

## 载入待测图片

```python
img_path = 'tests/data/rec_toy_dataset/imgs/1036169.jpg'
```

```python
img_bgr = cv2.imread(img_path)
```

```python
plt.imshow(img_bgr[:,:,::-1])
plt.show()
```

## 执行预测

```python
result = infer(img_path, save_vis=True, return_vis=True)
```

## 解析预测结果

```python
result['predictions']
```

```python
plt.imshow(result['visualization'][0])
plt.show()
```
