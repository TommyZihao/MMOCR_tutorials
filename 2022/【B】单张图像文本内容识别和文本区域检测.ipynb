## 导入工具包

```python
# 导入 opencv-python
import cv2

# 导入可视化工具包 matplotlib，并让绘制的图像嵌入在 notebook 中
import matplotlib.pyplot as plt
%matplotlib inline

# 定义可视化图像函数，输入图像路径，可视化图像
def show_img_from_path(img_path):
    '''opencv 读入图像，matplotlib 可视化格式为 RGB，因此需将 BGR 转 RGB，最后可视化出来'''
    img = cv2.imread(img_path)
    img_RGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(img_RGB)
    plt.show()
    
# 定义可视化图像函数，输入图像 array，可视化图像
def show_img_from_array(img):
    '''输入 array，matplotlib 可视化格式为 RGB，因此需将 BGR 转 RGB，最后可视化出来'''
    img_RGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(img_RGB)
    plt.show()
```

## 文本区域检测

可供使用模型：https://mmocr.readthedocs.io/zh_CN/latest/demo.html#id6

### 文本区域检测的命令行调用

```python
!python mmocr/utils/ocr.py \
        demo/demo_text_det.jpg \
        --det TextSnake \
        --recog None \
        --output outputs/out_demo_text_det.jpg
```

```python
show_img_from_path('outputs/out_demo_text_det.jpg')
```

### 文本区域检测的 Python API调用

```python
# 加载预训练文本区域检测器
detector = MMOCR(det='TextSnake',  # 文本检测算法，这里指定为 TextSnake，也可替换为其他 MMOCR 支持的文本区域检测算法
                 recog=None,       # 文本识别算法，这里不指定，也可替换为其他 MMOCR 支持的文本识别算法
                 device='cpu')     # 指定运算设备为 cpu 或 cuda
```

```python
# 指定输入图像路径与输出图像路径，执行预测
result = detector.readtext('demo/demo_text_det.jpg', output='outputs/demo_text_det_pred.jpg')
```

```python
# 查看生成的效果图
show_img_from_path('outputs/demo_text_det_pred.jpg')
```

### 解析 result 中包含的文本区域坐标、置信度信息

```python
# result[0]['boundary_result']
```

```python
# 文本区域总个数
len(result[0]['boundary_result'])
```

```python
idx = 1

# 索引为idx的文本区域
len(result[0]['boundary_result'][idx])
```

`result[0]['boundary_result'][idx]` 列表中包含了索引为 idx 的文本区域中，每个点的横纵坐标，以及该区域置信度。

```python
# 列表前面为该文本区域每个点的横纵坐标
result[0]['boundary_result'][idx][:10]
```

```python
# 列表最后一位为文本区域置信度
result[0]['boundary_result'][idx][-1]
```

## 文本内容识别

### 文本内容识别的命令行调用

```python
!python mmocr/utils/ocr.py \
        demo/demo_text_recog.jpg \
        --recog SAR \
        --det None \
        --output outputs/out_demo_text_recog.jpg
```

```python
show_img_from_path('outputs/out_demo_text_recog.jpg')
```

### 文本内容识别的 Python API 调用

```python
# 加载预训练文本内容识别器
recognizer = MMOCR(det=None,      # 文本区域检测算法，这里不指定
                   recog='SAR',   # 文本识别算法，这里指定为 'SAR'，也可替换为其他 MMOCR 支持的文本识别算法
                   device='cpu')  # 指定运算设备为 cpu 或 cuda
```

```python
# 指定输入和输出图像路径，运行文本内容识别算法，获取结果和置信度
result = recognizer.readtext('demo/demo_text_recog.jpg', output='outputs/demo_text_recog_pred.jpg')
```

效果很差，为什么呢？

原来，文本内容识别算法，输入的图像一般是将文字区域裁剪出来的小图块，而不能直接输入全图。否则会像上图一样效果不佳。

应该先将全图输入至文本区域检测器，定位并裁剪出文本区域，再将文本区域的小块图像输入文本内容识别器，获得文本内容。
